```python
import numpy as np
import pandas as pd
import plotly.graph_objects as go
import plotly.express as px
df = pd.read_csv('/content/dkikepadatankelurahan2013.csv')
df
```

```python
#membuat scatter plot
import plotly.express as px
fig = px.scatter(df,
                 x = "NAMA KELURAHAN",
                 y = "LUAS WILAYAH (KM2)")
fig.update_layout(
    title = 'kepadatan penduduk',
    title_font_color = 'blue',
    title_font_size = 30,

    xaxis = dict(
        title ='Luas wilayah',
        titlefont_size = 16,
        tickfont_size = 14))

fig.show()
```
